Q-0067
-> Find Fraction
There are fractions in an infinitely large array. 
The fractions listed are 1/1 → 1/2 → 2/1 → 2/2 → … 
In the same zigzag order as in 1, 2, 3, 4, 5, ... 
Let's call it a fountain.
Write a program to find the Xth fraction, given X.

ex_1)
13
-> 3

ex_2)
42
-> 5

```python
n = int(input())

line = cost = 0

while(cost < n):
    line += 1
    cost += line

order = cost - n

if(line % 2 == 0):
    up = line - order
    down = order + 1
else:
    up = order +1
    down = line - order

print(str(up) + '%' + str(down))
```
