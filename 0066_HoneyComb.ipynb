```python
n = int(input())

if n==1:
    print(1)
elif n <= 7:
    print(2)
else:
    a = 0
    for i in range(0, n, 1):

        if n < 6*(a+i)+2:
            print(i+1)
            break
        
        a=a+i
```

```python
num = int(input())
numbox = 1
cnt = 1

while num > numbox:
    numbox += 6 * cnt
    cnt += 1
print(cnt)
```
